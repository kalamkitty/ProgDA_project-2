# Overview of Breast Cancer
***

Breast cancer is the abnormal growth of cells in the breast, causing a tumour. It is the most common form of cancer diagnosed in women.  Breast cancer is more common amongst women but men can also be diagnosed (Mayo Clinic, 2022).

According to the Irish Cancer Society, every year, there are more than 3,500 women in Ireland get diagnosed with breast cancer, more common in women over 50 (Irish Cancer Society,_).

There are many risk factors for breast cancer such as lack of activity, alcohol consumption and hormonal imbalance. Breast cancer can be detected early by self examination and regular mammograms, these scans show any abnormal areas of the breast, detecting any tumours (Cancer.org, 2022).

# Dataset Background
***

The Wisconsin Breast Cancer (original) Dataset was compiled by three individuals, Wolberg, Street and Mangasarian at the University of Wisconsin in 1996. The dataset consists of records of measurements for 699 cases/samples of breast cell nuclei based on 10 attributes. The data was collected by Dr. Wolberg's in clinical cases by a non surgical method called Fine Needle Biopsy which extracts a small amount of cells from the tumor.


The attributes of the dataset are as follows;

- ID number/Sample number
- Clump Thickness (1-10)
- Uniformity of Cell Size (1-10)
- Uniformity of Cell Shape (1-10)
- Marginal Adhesion (1-10)
- Single Epithelial Cell Size (1-10)
- Bare Nuclei (1-10)
- Bland Chromatin (1-10)
- Normal Nucleoli (1-10)
- Mitoses (1-10)
- Class (2 for Benign, 4 for Malignant)

There are 16 instances in the data with a missing attribute value, denoted by "?".
I have downloaded the dataset from the UCI database [Link](https://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+(original)

#### Random Forests

The Random Forest Classifier is one of the most popular classification methods, it consists of large ensemble of individual decision trees, trained with the "bagging" method to accurately predict results. Each individual decision tree results in a class prediction and the most common prediction becomes the random forest model's prediction (Yu,2019).

Bagging or bootstrapping creates random samples of data from the dataset and replace data so that each individual decision tree have a different outcome. Rather than testing all of the data, decision trees makes decisions on a selection of data. Each tree us trained on different data with different features to make descisions, providing a buffer for any errors and incorrect predictions (Palmer, 2019).



Random forest works well when there is low correlation between models. The large number of decision trees cancel out each other's individual errors.....

```python
# importing libraries
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV data and form a dataframe.
df = pd.read_csv("breast-cancer-wisconsin.data", header= None)
df
```

```python
# Adding headings to columns
df.columns = ["Case number", "Clump Thickness", "Uniformity of Cell Size", "Puniformity of Cell Shape", "Marginal Adhesion", "Single Epithelial Cell Size", "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli", "Mitoses", "Class"]
df
```

```python
# Count number of 2/4 from class column
df ["Class"].value_counts()
```

```python
# Remove column of case/sample number as it has no use in stat analysis.
df = df.drop(["Case number"], axis=1).drop(columns=df.columns[-1], axis=1)
df
```

```python
df.describe()
```

The last attribute (mitoses column is categorical rather than numerical, therefore I will use a pie chart to represent how many percentage of tumours are benign or malignant. From the .value_counts(), we can see that 458 cases are benign and 241 cases are malignant.

```python
# https://www.w3schools.com/python/matplotlib_pie_charts.asp
# Create a pie chart using the information from class column.

y= np.array([458,241])
myexplode = [.1,0]  
mylabels = ["Benign", "Malignant"]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)  
plt.legend(title = "Tumour Class")  
plt.show() 
```

To cleanse the data, I must check for any missing data. Ase mentioned in the dataset background, there are values labeled "?" in the dataset. I will replace "?" with 0 for visualisation purposes later.

```python
# https://stackoverflow.com/questions/17071871/how-do-i-select-rows-from-a-dataframe-based-on-column-values - boolean indexing
# Check if there are any "?" in each column.
columns = df.columns
for x in columns:
    print("Number of missing values in ", x, df[df[x] == '?'].shape[0])
```

```python
# https://www.geeksforgeeks.org/python-pandas-dataframe-replace/
# https://stackoverflow.com/questions/21287624/convert-pandas-column-containing-nans-to-dtype-int
df2 = df.replace(to_replace="?",value="0").astype(int)
df2
```

```python
df2.describe()
```

### Visualisation of data

The below histograms show the distribution of data for each different variable.

```python
# Plot histograms using seaborn with df2, where "?" are replaced by 0.
fig, ax = plt.subplots(4,2, figsize=(50,50))
sns.set(font_scale = 3)
sns.histplot(data= df2, x= "Clump Thickness", ax = ax[0,0],kde= True, color= "navy")
sns.histplot(data= df2, x= "Uniformity of Cell Size", ax = ax[0,1], kde = True, color= "red")
sns.histplot(data= df2, x= "Puniformity of Cell Shape", ax = ax[1,0], kde= True, color= "green")
sns.histplot(data= df2, x= "Marginal Adhesion", ax = ax[1,1], kde= True, color= "maroon")
sns.histplot(data= df2, x= "Bare Nuclei", ax = ax[2,0], kde= True, color= "blue")
sns.histplot(data= df2, x= "Bland Chromatin", ax = ax[2,1], kde= True, color= "purple")
sns.histplot(data= df2, x= "Normal Nucleoli",ax = ax[3,0], kde= True, color= "orange")
sns.histplot(data= df2, x= "Mitoses",ax = ax[3,1], kde= True, color= "teal")
```

## Correlation

The correlation matrix show the correlation between each attributes, how they relate to each other.
Correlation coefficient between .50 and 1 is considered strong correlation. Moderate at .30 to 0.49 and low correlation when the coeficient is below 0.29 (Calkins, 2005).

```python
df2.corr()
```

From the correlation matrix, the variables with the highest correlation are Uniformity of Cell Size and Puniformity of Cell Shape (0.906882). High correlation does not mean one attribute cause the other. The sample size can have a affect on correlation with the larger the sample, the more stable the obtained correlation.

As there are so many variables, I will use Seaborn's heatmap to visualise the correlation betweent the variables. 

```python
# https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e
plt.figure(figsize= (30, 10))

# Plot heat map, minimum coefficient is 0 and highest 1, annot true to add the coefficient onto the rectangles.
heatmap = sns.heatmap(df2.corr(), vmin= 0, vmax= 1,annot= True)
heatmap.set_title('Correlation Heatmap', fontdict= {'fontsize':40}, pad= 12);
```

From looking at the heatmap, we can see that the higher the correlation between attributes, the lighter the colour.
