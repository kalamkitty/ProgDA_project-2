# Overview of Breast Cancer
***

Breast cancer is the abnormal growth of cells in the breast, causing a tumour. It is the most common form of cancer diagnosed in women.  Breast cancer is more common amongst women but men can also be diagnosed (Mayo Clinic, 2022).

# Dataset Background
***

The Wisconsin Breast Cancer (original) Dataset was compiled by three individuals, Wolberg, Street and Mangasarian at the University of Wisconsin in 1996. The dataset consists of records of measurements for 699 cases/samples of breast cell nuclei based on 10 attributes. The data was collected by Dr. Wolberg's in clinical cases by a non surgical method called Fine Needle Biopsy which extracts a small amount of cells from the tumor.


The attributes of the dataset are as follows;

- ID number/Sample number
- Clump Thickness (1-10)
- Uniformity of Cell Size (1-10)
- Puniformity of Cell Shape (1-10)
- Marginal Adhesion (1-10)
- Single Epithelial Cell Size (1-10)
- Bare Nuclei (1-10)
- Bland Chromatin (1-10)
- Normal Nucleoli (1-10)
- Mitoses (1-10)
- Class (2 for Benign, 4 for Malignant)

There are 16 instances in the data with a missing attribute value, denoted by "?".

## Classifiers
Scikit-Learn is a python library developed for the purpose of classification of data and machine learning tasks. A classifier is a machine learning algorithm that organises an input element based on features into classes (Nelson, 2022).

The true negative rate (TNR) was also used to describe how well the classifiers worked in predicting a result. TNR is the probability of a negative test for a true negative will test negative.

#### Random Forests

The Random Forest Classifier is one of the most popular classification methods, it consists of large ensemble of individual decision trees, trained with the "bagging" method to accurately predict results. Each individual decision tree results in a class prediction and the most common prediction becomes the random forest model's prediction (Yu,2019).

Bagging or bootstrapping creates random samples of data from the dataset and replace data so that each individual decision tree have a different outcome. Rather than testing all of the data, decision trees makes decisions on a selection of data. Each tree us trained on different data with different features to make descisions, providing a buffer for any errors and incorrect predictions (Palmer, 2019).



Random forest works well when there is low correlation between models. The large number of decision trees cancel out each other's individual errors. It's takes more time than other classifiers but it predicts with a high level of accracy, works with datasets with missing data. No overfitting therefore can be used on new data.


In short, these are the adavantages and disadvantages of using random forest:

*Advantages*: Versatile, no overfitting and highly accruate

*Disadvantages*: Slower results and low interpretability


In the journal by Vig mentioned above, the success rate are as follows:

Random Forest (100) - **95.64%**

Random Forest (10) -  **90.13%**

#### Artificial Neural Networks

Artificial neural networks (ANN) is another machine learning model that performs prediction and classification tasks. It consists of artificial neurons which are a copy of brain neurons, these neurons connect in a network to perform tasks, the link between neurons is called weight.


ANN has three layers, an input layer, hidden layer and an output layer.
The neurons in the input layer sends information to the hidden layer. The hidden layer performs the neccessary tasks then transfer this as output to the output layer. The weight value determines how well the neural network can learn.

This model can perform more than one task simutaniously. If one neuron or weight does not work, this will not affect the system, although accurate, the main disadvantage of ANN is that the model does not give a detailed explanation for it's output (Mesquita, 2022).

Using this classifier, the success rate was **92.44%**

#### Support Vector Machines

Support Vector Machines (SVM) are another type of supervised machine learning algorithm used for classificaion tasks. SVM creates a "hyperplane" in space which acts as a boundary to separate the data into different classes. The main goal of SVM is to find a maximum marginal hyperplane that best divide dataset into classes.



The margin, as seen in the figure above, is the gap of the two closest points between the classes. A large margin is considered a good margin.

Like random forests, SVM provides fast and accruate predictions. However it is only used for smaller datasets as this classifer is more time consuming than others (Navlani, 2019).

Using this classifier, the success rate in the study are as follows:

SVM (linear) **78.45%**

SVM (rbf)  **93.63%**

SVM (quadratic) **72.93%**

#### Naïve Bayes Classifier
Naïve Bayes (NB) is an classification algorithm which is based on Bayes theorem. Bayes theorum finds the probability of a hypothesis with given evidence. This classifier assume that value of one feature is independend than another feature, regardless of any similarities/correlations. 

The Bayes theorem:

P(B|A) = Probability of B **if** A is true.

P(A) = Probability of hypothesis before observing evidence

P(B) = Probability of evidence

This classifier provides oversimplifies output but it is used in many real-world situations such as credit scoring and spam filtering. Only small amount of data is needed to train NB to make a prediction, this classifier is also very fast compared to more complicated methods like artificial neural networks (Geeksforgeels.org, 2023).

Using this classifier, the success rate in the study was **65.27%**

# Training Classifiers

The methods used in the Comparative Analysis of Different Classifiers for the Wisconsin Breast Cancer Dataset, split the data into 70% training set, 15% testing set and 15% validation set to predict whether a tumour is benign or malignant.

It is important to split dataset in machine learning to avoid overfitting, this is where a model that is training data so well that it negatively impacts the performance of the model of new data, any minor fluctuations are considered in the training data but generalises any new data (Brownlee, 2019).



```python
# Import packages
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# https://realpython.com/train-test-split-python-data/
# Splitting dataset into x features, dropping Class column and set as y target
x = df2.drop("Class", axis =1)
y = df2.Class
# Splitting sample of 20% of data as test data and 80% for training data.Random state used as a set value throughout to gurantee same sequence of random numbers each time.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Random Forest

```python
# Random Forest
# https://data36.com/random-forest-in-python/
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html
# Defining model, 50 number of trees, using same random state.
rf1 = RandomForestClassifier(n_estimators= 50, max_features= "auto", random_state= 42)

# Traing rf model.
rf1.fit(x_train, y_train)
rf1_predictions = rf1.predict(x_test)
rf1_predictions
```

```python
# Random forest with 100 number of trees
rf2 = RandomForestClassifier(n_estimators= 100, max_features= "auto", random_state= 42)
rf2.fit(x_train, y_train)
rf2_predictions = rf2.predict(x_test)
rf2_predictions
```

```python
# Accuracy score
# https://www.javatpoint.com/accuracy_score-in-sklearn
y_pred = rf1.predict(x_test)  
rf1score = accuracy_score(y_test, y_pred)*100
print(rf1score)
```

```python
y_pred = rf2.predict(x_test)  
rf2score = accuracy_score(y_test, y_pred)*100
print(rf2score) 
```

### Support Vector Machines

```python
# https://pythonprogramming.net/linear-svc-example-scikit-learn-svm-python/
# Define Classifier.
svm = svm.SVC(random_state= 42)
svm.fit(x_train, y_train)
svm_predictions = svm.predict(x_test)
svm_predictions
```

```python
y_pred = svm.predict(x_test)  
svmscore = accuracy_score(y_test, y_pred)*100
print(svmscore) 
```

## Naive Bayes

```python
# https://www.analyticsvidhya.com/blog/2021/01/a-guide-to-the-naive-bayes-algorithm/
nb = GaussianNB()
nb.fit(x_train, y_train)
nb_predictions = nb.predict(x_test)
nb_predictions
```

```python
y_pred = nb.predict(x_test)  
nbscore = accuracy_score(y_test, y_pred)*100
print(nbscore) 
```

## K-Nearest Neighbors

```python
# https://www.geeksforgeeks.org/k-nearest-neighbor-algorithm-in-python/
knn = KNeighborsClassifier(n_neighbors= 7)
knn.fit(x_train, y_train)
knn_predictions = knn.predict(x_test)
knn_predictions
```

```python
y_pred = knn.predict(x_test)  
knnscore = accuracy_score(y_test, y_pred)*100
print(knnscore) 
```

## Decision trees

```python
# https://www.w3schools.com/python/python_ml_decision_tree.asp
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x_train, y_train)
dtree_predictions = dtree.predict(x_test)
dtree_predictions
```

# Data Synthesis

The reasoning for data synthesis is that it can be used to train classifiers for more accruate results. The Wisconsin Breast Cancer is a small dataset with 699 samples. It is also health data which can is expensive to obtain and with GDPR and privacy lawss, health data is sensitive/rare data and there is a question whether it should be released to be used by the general public.

Synethic data "minmics" the data thats already in dataset, it is non randomised and is used to further train machine learning models.

There are two ways to synthesize data,

1. Generative Models - Generative adversarial networks are models for synthesizing data. Made up of two sub-models: generator and discrimiator. As the name suggests, generator generates the new data and the later verifies the data and the latter determines if the data looks real or fake. As new data points are generated, the more realistic they look to the discrimiator, the more the generator will create more data points.


2. Conventional Methods - Using software and tools to generate data or partnering with a third party. These methods ccan be free and data will be suitable for testing but partnering with another company will require networking. Partnership can be very valuable as the other party may be more experienced/specialise in one specific type of data (altexoft.com, 2022).
